```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. Load Dataset
We will use the famous **Iris dataset** (built into seaborn).

```python
# Load dataset
df = sns.load_dataset("iris")

# Show first few rows
df.head()
```

## 2. Dataset Overview
- Shape (rows, columns)
- Column data types
- Null values

```python
print("Shape of dataset:", df.shape)
print("\nDataset Info:\n")
print(df.info())
print("\nMissing Values:\n", df.isnull().sum())
```

## 3. Descriptive Statistics
We calculate mean, median, standard deviation, etc.

```python
df.describe()
```

## 4. Univariate Analysis
Check distributions of single variables.

```python
# Histogram for sepal_length
plt.figure(figsize=(7,5))
sns.histplot(df["sepal_length"], bins=20, kde=True, color="skyblue")
plt.title("Sepal Length Distribution")
plt.show()
```

## 5. Multivariate Analysis
Explore relationships between variables.

```python
# Scatterplot between sepal_length and petal_length
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x="sepal_length", y="petal_length", hue="species", palette="Set2")
plt.title("Sepal Length vs Petal Length")
plt.show()

# Pairplot (correlation across all features)
sns.pairplot(df, hue="species", palette="Set1")
plt.show()
```

## 6. Correlation Heatmap
Check how numeric features are related.

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=[np.number])

# Correlation Heatmap
plt.figure(figsize=(6,4))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap (Numeric Features Only)")
plt.show()
```

## 7. Key Insights
- Petal length & petal width strongly correlate with species type.
- Sepal measurements overlap more across species.
- Dataset is clean (no missing values).
