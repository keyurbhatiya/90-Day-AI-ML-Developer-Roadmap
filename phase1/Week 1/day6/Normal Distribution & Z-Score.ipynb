# Day 6: Normal Distribution & Z-Score
In this notebook, we will learn:
1. What is Normal Distribution
2. Visualizing Normal Distribution
3. What is Z-Score
4. How to calculate and interpret Z-Score

```python
# Import Libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm, zscore
```

## 1. Normal Distribution
- A type of continuous probability distribution.
- Symmetrical, bell-shaped curve.
- Mean = Median = Mode.

```python
# Generate normally distributed data
np.random.seed(42)  # for reproducibility
data = np.random.normal(loc=50, scale=10, size=1000)  
# loc = mean, scale = standard deviation

# Plot histogram with KDE
plt.figure(figsize=(8,5))
sns.histplot(data, bins=30, kde=True, color="skyblue")
plt.title("Normal Distribution (mean=50, std=10)")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

## 2. Z-Score
- Z-score tells us **how many standard deviations** a data point is away from the mean.
- Formula:

\[
Z = \frac{(X - \mu)}{\sigma}
\]

Where:
- \(X\) = data point  
- \(\mu\) = mean  
- \(\sigma\) = standard deviation

```python
# Calculate Mean and Standard Deviation
mean = np.mean(data)
std = np.std(data)

print("Mean:", mean)
print("Standard Deviation:", std)

# Example: Z-score of a value X=70
X = 70
z = (X - mean) / std
print(f"Z-score for X={X}:", z)
```

## 3. Z-Score with Scipy
We can also calculate Z-scores for the entire dataset using `scipy.stats.zscore`.


```python
# Compute Z-scores for all data points
z_scores = zscore(data)

print("First 10 Z-scores:\n", z_scores[:10])

# Plot Z-score distribution
plt.figure(figsize=(8,5))
sns.histplot(z_scores, bins=30, kde=True, color="lightgreen")
plt.title("Z-score Distribution (Standard Normal)")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()
```

## 4. Why Z-scores are Important?
- Z-scores allow comparison between datasets with different scales.
- Used to detect **outliers**:
  - If |Z| > 3  considered an outlier.
- Used in hypothesis testing & probability calculations.
